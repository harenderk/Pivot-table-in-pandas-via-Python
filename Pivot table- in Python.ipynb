```python
import pandas as pd
import numpy as np
```

```python
# importing file in python
data = pd.read_excel('superstore_dataset2011-2015.xlsx')
```

```python
data.head() # viewing file
```

```python
c= list(data.columns) # showing columns
print(c)
```

```python
# creating pivot table by selecting desired columns
pivot = pd.pivot_table(data,index=['Segment','Order ID','Customer ID','Customer Name'],values=['Sales'],aggfunc=np.sum)
```

```python
pivot.head()
```

```python
# saving results.
pivot.to_excel('pivot.xlsx')
```
